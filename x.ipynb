```python
a = np.int64(2)
x = a * a
```

```python
cnt = np.int64(2)
dic = {}
```

```python
c = np.int64(1000000000)
```

```python
while x % c != a:
    x *= a
    x %= c
    cnt += 1
    if x in dic.keys():
        print ("Answer found", cnt, dic[x])
        break
    dic[x] = cnt
    if cnt % 10000 == 0:
        print(cnt)
    if cnt > (2*c):
        break
```
